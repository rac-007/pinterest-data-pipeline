### Loading cleaned user data

```
json_file_path = "/mnt/my-json-files/cleaned_user_json"
df_user = spark.read.json(json_file_path)
df_user = df_user.select("ind", "user_name","age","date_joined")
display(df_user)
```

```
from pyspark.sql.functions import count, col, row_number
from pyspark.sql.window import Window

# Step 1: Join df_pin and df_geo on 'ind'
df_pin_geo = df_pin.join(df_geo, on="ind")

# Step 2: Group by 'country' and 'category', and count occurrences
df_grouped = df_pin_geo.groupBy("country", "category") \
    .agg(count("category").alias("category_count"))

# Step 3: Use a window function to rank categories by their count in each country
window_spec = Window.partitionBy("country").orderBy(col("category_count").desc())

df_ranked = df_grouped.withColumn("rank", row_number().over(window_spec))

# Step 4: Filter to get the most popular category per country
df_most_popular = df_ranked.filter(col("rank") == 1).select("country", "category", "category_count")

# Display the result
df_most_popular.show()
```

### Find the most popular category for different age group
#### What is the most popular category people post to based on the following age groups:
- 18-24
- 25-35
- 36-50
- +50
#### Your query should return a DataFrame that contains the following columns:
- age_group, a new column based on the original age column
- category
- category_count, a new column containing the desired query output

```
from pyspark.sql.functions import when, col, count
from pyspark.sql.window import Window
# Step 1: Define age groups
df_user_with_age_group = df_user.withColumn(
    "age_group",
    when((col("age") >= 18) & (col("age") <= 24), "18-24")
    .when((col("age") >= 25) & (col("age") <= 35), "25-35")
    .when((col("age") >= 36) & (col("age") <= 50), "36-50")
    .when(col("age") > 50, "+50")
    .otherwise("Unknown")
)
df_user_with_age_group.show()
```

```
# Step 2: Join df_user_with_age_group with df_pin on 'ind' to include category information
df_combined = df_pin.join(df_user_with_age_group, on="ind")
```

```
# Step 3: Group by 'age_group' and 'category', and count the occurrences
df_grouped = df_combined.groupBy("age_group", "category") \
    .agg(count("category").alias("category_count"))
display(df_grouped)
```

```
# Step 4: Use a window function to rank categories by their count for each age group
window_spec = Window.partitionBy("age_group").orderBy(col("category_count").desc())
df_ranked = df_grouped.withColumn("rank", row_number().over(window_spec))
df_ranked.show()
```

```
# Step 5: Filter to get the most popular category per age group
df_most_popular_by_age_group = df_ranked.filter(col("rank") == 1).select("age_group", "category", "category_count")
```

```
# Display the result
df_most_popular_by_age_group.show()
```

### Find the median follower count for different age group
#### What is the median follower count for users in the following age groups:
- 18-24
- 25-35
- 36-50
- +50
#### Your query should return a DataFrame that contains the following columns:
- age_group, a new column based on the original age column
- median_follower_count, a new column containing the desired query output

```

df_user_with_age_group.show()
df_combined = df_pin.join(df_user_with_age_group, on="ind")
```

```
from pyspark.sql.functions import when, col, expr
from pyspark.sql.window import Window
import pyspark.sql.functions as F
# Calculate median follower count for each age group
window_spec = Window.partitionBy("age_group").orderBy("follower_count")
df_combined_with_rank = df_combined.withColumn(
    "rank", F.row_number().over(window_spec)
)
# expr() function takes SQL expression as a string argument, executes the expression, and returns a PySpark Column type object.
# Calculate the median
df_median = df_combined_with_rank.groupBy("age_group").agg(
    expr("percentile_approx(follower_count, 0.5)").alias("median_follower_count")
)

# Display the result
df_median.show()
```

### Find the median follower count of users based on their joining year
#### Find the median follower count of users have joined between 2015 and 2020.
#### Your query should return a DataFrame that contains the following columns:
- <b>post_year</b>, a new column that contains only the year from the timestamp column
- <b>median_follower_count</b>, a new column containing the desired query output

```

#  Filter the DataFrame to include only users who joined between 2015 and 2020
df_filtered = df_user_with_year.filter((col("post_year") >= 2015) & (col("post_year") <= 2020))
#df_filtered.show()
# Join df_filtered with df_pin to include follower_count information
df_combined = df_pin.join(df_filtered, on="ind")
# Calculate the median follower count for each year
# Define a window specification to calculate the rank within each year
window_spec = Window.partitionBy("post_year").orderBy("follower_count")

# Add a rank column to calculate the median
df_combined_with_rank = df_combined.withColumn(
    "rank", F.row_number().over(window_spec)
)
# Calculate the median follower count for each year using percentile_approx
df_median_follower_count = df_combined_with_rank.groupBy("post_year").agg(
    expr("percentile_approx(follower_count, 0.5)").alias("median_follower_count")
)

# Display the result
df_median_follower_count.show()
```

### Find the median follower count of users based on their joining year and age group
#### Find the median follower count of users that have joined between 2015 and 2020, based on which age group they are part of.
#### Your query should return a DataFrame that contains the following columns:
- <b>age_group</b>, a new column based on the original age column
- <b>post_year</b>, a new column that contains only the year from the timestamp column
- <b>median_follower_count</b>, a new column containing the desired query output


```
from pyspark.sql import functions as F

# Step 1: Extract the year and create age groups
df_with_year_age_group = df_user.withColumn("post_year", F.year("date_joined")).withColumn(
    "age_group",
    F.when((F.col("age") >= 18) & (F.col("age") <= 24), "18-24")
    .when((F.col("age") >= 25) & (F.col("age") <= 35), "25-35")
    .when((F.col("age") >= 36) & (F.col("age") <= 50), "36-50")
    .when(F.col("age") > 50, "+50")
)

# Ensure the follower_count column exists
if 'follower_count' in df_user.columns:
    # Step 2: Filter the data for the years 2015 to 2020
    df_filtered = df_with_year_age_group.filter(
        (F.col("post_year") >= 2015) & (F.col("post_year") <= 2020)
    )

    # Step 3: Group by age_group and post_year, and calculate the median follower count
    df_median_follower_count = df_filtered.groupBy("age_group", "post_year").agg(
        F.expr("percentile_approx(follower_count, 0.5)").alias("median_follower_count")
    )

    # Display the result
    display(df_median_follower_count)
else:
    print("Column 'follower_count' does not exist in the DataFrame.")
```
